```python
import gradio as gr #fur UI
from transformers import pipeline #fur Transformer pipeline
```

2. Load Up Pipeline

```python
translation_pipeline = pipeline('translation_en_to_de')
```

```python
results= translation_pipeline('I love ice')
```

```python
results[0]['translation_text']
```

**Create Gradio Function and Interface**

```python
def translate_lang(text_to_translate):
  results = translation_pipeline(text_to_translate)
  return results[0]['translation_text']
```

```python
translate_lang('i have one or two')
```

```python
interface = gr.Interface(fn=translate_lang,
                         inputs=gr.components.Textbox(lines=2, placeholder='Text to translate'),
                        outputs='text')
```

```python
interface.launch()
```
